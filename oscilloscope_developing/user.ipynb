### Start the test!
* If you prepare to do the test, then click the start test. After clicking the "Start Test!", you will see the following page

```python
Image("4.png")
```

* In choose a channel, you need to make sure you choose the right channel that you are using for measurement or you cannot correctly control the oscilloscope
* In impedance section, choose the coupling and impedance you are required for the experiment
* In type of slope section, choose the slope you are required for the experiment
* auto scale button is used for doing auto scale, the oscilloscope will automatically scale the waveform
* There are four sliders which can be used for manually adjust the scale or offset of the waveform, you can adjust by yourself. **Notice: there maybe some delay between the device and the control pannel, when you change the slider, you should wait a little bit to see if there is a change in the display**
* In the depth of memory section, you are allowed to define the depth of memory (number of points), the automatic button will set the depth of memory to 10 Mpts
* In the sampling rate section, you are allowed to define the sampling rate, the automatic button will set the sampling rate to 2 GSa/s
* When the average button is activated, oscilloscope acquires multiple data values for each time bucket, and averages them
* Click the "Show the waveform", you want to check the waveform. **Notice: you may need to scale the waveform by yourself in the picture window(very easy to do)**
* Click the "Download the data measurement" if you want to download some main measurements. Measurements include rising time, falling time, frequency, period, amplitude, pulse width, duty cycle.
    * You need to specify the path of csv file you want to download, the default download path is where the software program is. You do not need to specify the type of the file (csv file).
* Click the "Download the waveform data" if you want to download the data points of the waveform.
    * You need to specify the path of csv file you want to download, the default download path is where the software program is. You do not need to specify the type of the file (csv file). The file is very large!
* Click the "Download the waveform plot" if you want to download the waveform plot.
    * You need to specify the path of png file you want to donwload, the default download path is where the software program is. You do not need to specify the type of the file (png file).
* Click the "Download the screenshot" if you want to download the screenshot of the display of oscilloscope.
    * You need to specify the path of png file you want to donwload, the default download path is where the software program is. You do not need to specify the type of the file (png file).
